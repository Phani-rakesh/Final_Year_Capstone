# cleaning chunks and saving as parquet files

```python
import pandas as pd
import os
import glob

# ✅ Path to your original chunk CSV files in Drive
chunk_folder = "alzimers_problilities"   # 👈 change if different
chunk_files = sorted(glob.glob(os.path.join(chunk_folder, "*.csv")))

# ✅ Path to cleaned parquet output folder in Drive
cleaned_folder = "cleaned_chunks"
os.makedirs(cleaned_folder, exist_ok=True)

print(f"✅ Found {len(chunk_files)} chunk files")

for i, f in enumerate(chunk_files, 1):
    print(f"🧼 Cleaning chunk {i}/{len(chunk_files)} → {os.path.basename(f)}")

    # Load chunk
    df = pd.read_csv(f)

    # Strip spaces in column names
    df.columns = df.columns.str.strip()

    # Fill NaNs with 0.5
    df = df.fillna(0.5)

    # Round all CpG probability values to 6 decimal places
    numeric_cols = df.select_dtypes(include=['float64', 'float32', 'int64']).columns
    df[numeric_cols] = df[numeric_cols].round(6)

    # Save cleaned file to Google Drive as Parquet
    out_file_name = os.path.splitext(os.path.basename(f))[0] + ".parquet"
    out_file = os.path.join(cleaned_folder, out_file_name)
    df.to_parquet(out_file, index=False)

print("✅ All chunks cleaned and saved as Parquet in Google Drive:", cleaned_folder)
```

```python
# 1️⃣ Load labels CSV
labels_path = "alzheimers_balanced.csv'  # ✅ make sure this path is correct
labels_df = pd.read_csv(labels_path)
labels_df["sample_id"] = labels_df["sample_id"].astype(str).str.strip()

# 2️⃣ Load the same test file again
df = pd.read_parquet(test_file)

# 3️⃣ Merge labels
df = df.merge(labels_df[["sample_id", "label"]], on="sample_id", how="inner")

print("✅ After merging labels:", df.shape)
print(df[["sample_id", "label"]].head())
```

```python
# ✅ Extract target and CpG features
y = df['label'].values
X = df.drop(columns=['sample_id', 'label'], errors='ignore')

print("📝 Initial features shape:", X.shape)
print("🧠 Target shape:", y.shape)

# ⚠️ Remove constant (zero variance) CpGs
zero_var_cols = X.columns[X.std() == 0]
print(f"⚠️ Removing {len(zero_var_cols)} constant CpG features")
X = X.drop(columns=zero_var_cols)

print("✅ Final features shape after cleaning:", X.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy

# ✅ 1️⃣ Configure a lighter Random Forest for Boruta
rf = RandomForestClassifier(
    n_estimators=150,    # 🔸 Lower number of trees to save memory
    max_depth=10,       # 🔸 Limit tree depth → reduces model size drastically
    n_jobs=-1,          # Use all available CPU cores
    random_state=42
)

# ✅ 2️⃣ Initialize Boruta with safe settings
boruta_selector = BorutaPy(
    estimator=rf,
    n_estimators='auto',
    verbose=1,         # Moderate verbosity so it doesn't spam output too much
    random_state=42
)

# ✅ 3️⃣ Run Boruta
print("🚀 Running Boruta feature selection (safe mode)...")
boruta_selector.fit(X.values, y)
print("✅ Boruta feature selection completed safely ✅")
```

```python
import pandas as pd

# ✅ 1️⃣ Extract confirmed features
selected_mask = boruta_selector.support_
selected_features = X.columns[selected_mask]

# ✅ 2️⃣ Extract tentative features (optional, sometimes useful)
tentative_mask = boruta_selector.support_weak_
tentative_features = X.columns[tentative_mask]

print(f"🌿 Confirmed important CpGs: {len(selected_features)}")
print(f"🤔 Tentative CpGs: {len(tentative_features)}")
print("\n🧬 First 20 confirmed CpGs:\n", list(selected_features[:20]))

# ✅ 3️⃣ Save to CSV
selected_features_path = "boruta_selected_cpgs_chunk1_confirmed.csv"
tentative_features_path = "boruta_selected_cpgs_chunk1_tentative.csv"

pd.Series(selected_features).to_csv(selected_features_path, index=False)
pd.Series(tentative_features).to_csv(tentative_features_path, index=False)

print(f"\n💾 Confirmed CpGs saved to: {selected_features_path}")
print(f"💾 Tentative CpGs saved to: {tentative_features_path}")
```

```python
import pandas as pd
import glob, os
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy

# -----------------------------
# 1️⃣ Paths
# -----------------------------
cleaned_folder = "cleaned_chunks"
labels_path = "/alzheimers_balanced.csv"  # ✅ corrected path
out_folder = "/boruta_chunk_results_optimized"
os.makedirs(out_folder, exist_ok=True)

# -----------------------------
# 2️⃣ Load Labels Once
# -----------------------------
labels_df = pd.read_csv(labels_path)
labels_df["sample_id"] = labels_df["sample_id"].astype(str).str.strip()
if "label" not in labels_df.columns:
    raise ValueError("❌ Your labels CSV must contain a 'label' column (0/1).")

# -----------------------------
# 3️⃣ List All Chunks
# -----------------------------
parquet_files = sorted(glob.glob(f"{cleaned_folder}/*.parquet"))
print(f"✅ Found {len(parquet_files)} cleaned parquet files for Boruta")

# -----------------------------
# 4️⃣ RF Config
# -----------------------------
rf = RandomForestClassifier(
    n_estimators=150,
    max_depth=10,
    n_jobs=-1,
    random_state=42
)

# -----------------------------
# 5️⃣ RESUMABLE BORUTA LOOP
# -----------------------------
all_confirmed = []

for i, f in enumerate(parquet_files, 1):
    base_name = os.path.basename(f)
    confirmed_path = os.path.join(out_folder, f"boruta_confirmed_chunk_{i}.csv")
    tentative_path = os.path.join(out_folder, f"boruta_tentative_chunk_{i}.csv")

    # ⏸️ Skip already processed chunks
    if os.path.exists(confirmed_path) and os.path.exists(tentative_path):
        print(f"⏭️ Skipping chunk {i}/{len(parquet_files)} ({base_name}) — already processed.")
        continue

    print(f"\n🚀 [{i}/{len(parquet_files)}] Running Boruta on: {base_name}")

    try:
        df = pd.read_parquet(f)
        if "sample_id" not in df.columns:
            print("⚠️ No 'sample_id' in chunk, skipping.")
            continue

        df["sample_id"] = df["sample_id"].astype(str).str.strip()
        merged = df.merge(labels_df[["sample_id", "label"]], on="sample_id", how="inner")
        if merged.shape[0] == 0:
            print("⚠️ Merge produced 0 rows, skipping.")
            continue

        y = merged["label"].values
        X = merged.drop(columns=["sample_id", "label"], errors="ignore")

        # Drop constant features
        zero_var_cols = X.columns[X.std() == 0]
        if len(zero_var_cols) > 0:
            X = X.drop(columns=zero_var_cols)

        # Pre-filter top 2000 CpGs by variance
        if X.shape[1] > 2000:
            top_vars = X.var(axis=0).sort_values(ascending=False).head(2000).index
            X = X[top_vars]

        if X.shape[1] == 0:
            print("⚠️ No variable CpGs left after filtering, skipping.")
            continue

        boruta = BorutaPy(
            estimator=rf,
            n_estimators='auto',
            max_iter=20,
            verbose=1,
            random_state=42
        )
        boruta.fit(X.values, y)

        confirmed = X.columns[boruta.support_].tolist()
        tentative = X.columns[boruta.support_weak_].tolist()

        print(f"🌿 Confirmed: {len(confirmed)} | 🤔 Tentative: {len(tentative)}")

        pd.Series(confirmed).to_csv(confirmed_path, index=False)
        pd.Series(tentative).to_csv(tentative_path, index=False)

        all_confirmed.extend(confirmed)

    except Exception as e:
        print(f"❌ Error on chunk {i}: {e}")
        continue

# -----------------------------
# 6️⃣ UNION RESULTS
# -----------------------------
unique_confirmed = sorted(pd.unique(pd.Series(all_confirmed)))
union_path = os.path.join(out_folder, "boruta_confirmed_union_all_chunks.csv")
pd.Series(unique_confirmed).to_csv(union_path, index=False)

print(f"\n✅ Total unique confirmed CpGs across all chunks: {len(unique_confirmed)}")
print(f"💾 Saved union list to: {union_path}")
```

```python
import os, glob

# 📁 Folder containing cleaned parquet chunk files
chunks_folder = "cleaned_chunks"
chunk_files = sorted(glob.glob(f"{chunks_folder}/*.parquet"))
print(f"📂 Found {len(chunk_files)} chunk files to merge")

# 📌 Output file for incremental merged dataset
merged_output = "/alz_merged_selected_cpgs.parquet"

# 👉 Track resume progress
resume_index = 0
if os.path.exists(merged_output):
    print("🔄 Previous merged file found. Resuming merge...")
    merged_df = pd.read_parquet(merged_output)
    merged_ids = set(merged_df['sample_id'])
    resume_index = len(merged_df.columns) - 1  # minus sample_id
    print(f"✅ Resuming from {resume_index} features already merged.")
else:
    merged_df = None

# Merge loop
for i, file in enumerate(chunk_files, 1):
    print(f"\n🔸 [{i}/{len(chunk_files)}] Processing: {os.path.basename(file)}")
    df = pd.read_parquet(file)

    # Keep only sample_id + selected CpGs available in this chunk
    cols_to_keep = ["sample_id"] + [c for c in selected_cpgs if c in df.columns]
    df = df[cols_to_keep]

    # Skip if no relevant CpGs found
    if len(cols_to_keep) <= 1:
        print("⚠️ No selected CpGs in this chunk, skipping.")
        continue

    # Merge incrementally
    if merged_df is None:
        merged_df = df
    else:
        merged_df = pd.merge(merged_df, df, on="sample_id", how="outer")

    # Save progress after every 5 chunks
    if i % 5 == 0:
        print(f"💾 Saving intermediate merge after chunk {i}")
        merged_df.to_parquet(merged_output, index=False)

# Final save
merged_df.to_parquet(merged_output, index=False)
print(f"\n✅ Final merged dataset saved at: {merged_output}")
print("📐 Shape:", merged_df.shape)
```

```python
import pandas as pd

# -----------------------------
# 1️⃣ File paths
# -----------------------------
merged_file = "alz_merged_selected_cpgs.parquet"   # 👉 your merged CpG dataset
labels_file = "alzheimers_balanced.csv"         # 👉 label file (must have sample_id & label)

# -----------------------------
# 2️⃣ Load both files
# -----------------------------
df_merged = pd.read_parquet(merged_file)
labels_df = pd.read_csv(labels_file)

print("📊 Merged CpG file shape:", df_merged.shape)
print("🧪 Labels file shape:", labels_df.shape)

# -----------------------------
# 3️⃣ Clean up sample_id columns (to avoid mismatch)
# -----------------------------
df_merged['sample_id'] = df_merged['sample_id'].astype(str).str.strip()
labels_df['sample_id'] = labels_df['sample_id'].astype(str).str.strip()

# -----------------------------
# 4️⃣ Merge labels into merged CpG dataframe
# -----------------------------
df_labeled = df_merged.merge(labels_df[['sample_id', 'label']], on='sample_id', how='inner')

print("✅ After merging labels:", df_labeled.shape)
print(df_labeled[['sample_id', 'label']].head())

# -----------------------------
# 5️⃣ Save the labeled dataset for RFE & ML
# -----------------------------
output_path = "final_merged_labeled.parquet"
df_labeled.to_parquet(output_path, index=False)
print(f"💾 Labeled merged dataset saved to: {output_path}")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE

# -----------------------------
# 1️⃣ Load labeled dataset
# -----------------------------
file_path = "final_merged_labeled.parquet"
df = pd.read_parquet(file_path)
print("✅ Data loaded:", df.shape)

# -----------------------------
# 2️⃣ Separate features & target
# -----------------------------
y = df['label'].values
X = df.drop(columns=['sample_id', 'label'], errors='ignore')

print("📝 Features shape:", X.shape)
print("🧠 Target shape:", y.shape)

# -----------------------------
# 3️⃣ Configure Random Forest & RFE
# -----------------------------
num_features_to_select = 20  # Top 20 CpGs

rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    n_jobs=-1,
    random_state=42
)

rfe = RFE(
    estimator=rf_model,
    n_features_to_select=num_features_to_select,
    step=0.1   # eliminates 10% features each step — faster convergence
)

# -----------------------------
# 4️⃣ Run RFE
# -----------------------------
print("🚀 Running RFE... this may take a few minutes")
rfe.fit(X, y)
print("✅ RFE completed successfully!")

# -----------------------------
# 5️⃣ Extract selected features
# -----------------------------
selected_features = X.columns[rfe.support_].tolist()
print(f"🌿 Top {len(selected_features)} selected CpG sites:")
print(selected_features)

# -----------------------------
# 6️⃣ Save selected CpGs
# -----------------------------
output_path = "rfe_selected_top20_cpgs.csv"
pd.Series(selected_features).to_csv(output_path, index=False)
print(f"💾 Saved RFE-selected CpGs to: {output_path}")
```

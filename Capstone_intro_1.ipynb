```python
import pandas as pd

# Load train map
train_map = pd.read_csv("ai4bio_trainset\trainmap.csv")

# Filter only Alzheimer's disease and control
alz_data = train_map[train_map['disease'].isin(["Alzheimer's disease", "control"])]

# Separate positives and controls
alz_pos = alz_data[alz_data['disease'] == "Alzheimer's disease"]
alz_ctrl = alz_data[alz_data['disease'] == "control"]

# Randomly sample 500 each
alz_pos_sample = alz_pos.sample(n=500, random_state=42)
alz_ctrl_sample = alz_ctrl.sample(n=500, random_state=42)

# Combine
alz_balanced = pd.concat([alz_pos_sample, alz_ctrl_sample]).reset_index(drop=True)

# Shuffle
alz_balanced = alz_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Add ML label: 1 for Alzheimer's, 0 for control
alz_balanced['label'] = alz_balanced['disease'].apply(lambda x: 1 if x == "Alzheimer's disease" else 0)

# Save to Excel in your current working directory
alz_balanced.to_excel("alzheimers_balanced.xlsx", index=False)

print(alz_balanced['disease'].value_counts())
print("Final balanced shape:", alz_balanced.shape)
print("File saved as: alzheimers_balanced.xlsx")
```

```python
import pandas as pd

# Load your train map
train_map = pd.read_csv("ai4bio_trainset\trainmap.csv")

# Filter only Alzheimer's disease and controls
alzheimers_df = train_map[train_map["disease"].isin(["Alzheimer's disease", "control"])]

# Take 500 samples each
alzheimers_positive = alzheimers_df[alzheimers_df["disease"] == "Alzheimer's disease"].sample(n=500, random_state=42)
alzheimers_control = alzheimers_df[alzheimers_df["disease"] == "control"].sample(n=500, random_state=42)

# Combine them
balanced_df = pd.concat([alzheimers_positive, alzheimers_control])

# Save to Excel
output_file = "alzheimers_balanced.xlsx"
balanced_df.to_excel(output_file, index=False)

print(f"✅ File saved as {output_file}")
```
